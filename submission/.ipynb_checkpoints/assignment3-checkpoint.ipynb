```python
# Import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # Linear Regression model
```

## Process Data

```python
# Import training and testing datasets from ./data
df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
# Check for null values in training and testing datasets
print("\tTraining Data\n\n" + str(df.isnull().sum().sort_values(ascending=False)))
print("\n\n\tTesting Data\n\n" + str(df_test.isnull().sum().sort_values(ascending=False)))
```

```python
# Function used to process training and testing data
def process_data(df):
    df = df.select_dtypes(include=['int64','float64']) # New DataFrame created containing only dataset's numeric columns

    for i in range(len(df.columns)): # Iterate over the modified DataFrame containing only numeric columns
        feature_name = df.columns[i]
        if df[feature_name].isnull().sum() > 0: # Only modify columns that contain null values
            df[feature_name] = df[feature_name].fillna(df[feature_name].mean()) # Replace null values in column with column's mean
    return df
```

```python
# Process training and testing datasets
# This includes replacing all numeric columns' null values with that column's average
# as well as replacing the null values in non-numeric columns with the value 'NA'
df = process_data(df)
df_test = process_data(df_test)
```

```python
# Check for null values in training dataset
print("\tTraining Data\n\n" + str(df.isnull().sum().sort_values(ascending=False)))
print("\n\n\tTesting Data\n\n" + str(df_test.isnull().sum().sort_values(ascending=False)))
```

```python
# Function created to add 'Age' feature to training and testing datasets
def add_age_feature(df):
    df['Age'] = df['YrSold'] - df['YearBuilt']
    return df
```

```python
# Update datasets with new Age feature
df = add_age_feature(df)
df_test = add_age_feature(df_test)
```

```python
# Transform 'SalePrice' data using np.log() method and characterize as 'LogPrice' within training dataset
df['LogPrice'] = np.log(df['SalePrice'])

# Confirm that 'LogPrice' feature is populated with desired results
df['LogPrice'].head()
```

## Visualization of Data

```python
# Comparison of 'SalePrice' and 'LogPrice'
# 'SalePrice' is quite obviously skewed, while 'LogPrice' provides more balanced results 
plt.hist(df['SalePrice'], bins=20, edgecolor='white')
```

```python
# Histogram featuring 'LogPrice' statistics, more reliable than 'SalePrice'
plt.hist(df['LogPrice'], bins=20, edgecolor='white')
plt.title('LogPrice')
```

```python
# These features were modified after making the first submission and analyzing the data further
plt.hist(df['GrLivArea'], bins=20, edgecolor='white')
plt.title('GrLivArea')
plt.show()

plt.hist(df['TotalBsmtSF'], bins=20, edgecolor='white')
plt.title('TotalBsmtSF')
plt.show()

plt.hist(df['1stFlrSF'], bins=20, edgecolor='white')
plt.title('1stFlrSF')
plt.show()
```

```python
# Additional implementations (post-submission)
# Modify sweked features visualized above
df['GrLivArea'] = np.log(df['GrLivArea'].replace(0, 0.1)) # replace() function used due to division by 0 error
df_test['GrLivArea'] = np.log(df_test['GrLivArea'].replace(0, 0.1))

df['TotalBsmtSF'] = np.sqrt(df['TotalBsmtSF'].replace(0, 0.1))
df_test['TotalBsmtSF'] = np.sqrt(df_test['TotalBsmtSF'].replace(0, 0.1))

df['1stFlrSF'] = np.log(df['1stFlrSF'].replace(0, 0.1))
df_test['1stFlrSF'] = np.log(df_test['1stFlrSF'].replace(0, 0.1))
```

```python
# Visualization of updated feature values
plt.hist(df['GrLivArea'], bins=20, edgecolor='white')
plt.title('GrLivArea')
plt.show()

plt.hist(df['TotalBsmtSF'], bins=20, edgecolor='white')
plt.title('TotalBsmtSF')
plt.show()

plt.hist(df['1stFlrSF'], bins=20, edgecolor='white')
plt.title('1stFlrSF')
plt.show()
```

```python
# Generate correlation matrix to determine top features for training
corr = df.corr()
```

```python
# Select 'SalePrice' column from correlation matrix to determine how strongly correlated
# with 'SalePrice' the other features are
top_features = corr['SalePrice'].sort_values(ascending=False).index[2:11] # Correlations sorted in descending order
print("\n\tTop Features\n\n" + str(top_features))
```

```python
# Iterate over the top features in the dataset and visualize their relationship relative to 'SalePrice'
for feature in top_features:
    plt.scatter(df[feature], df['SalePrice'], alpha=0.5)
    plt.xlabel(str(feature))
    plt.ylabel('SalePrice')
    plt.show()
```

## Train and Test Linear Regression Model

```python
# Prepare features for Linear Regression model training
fit_col = list(top_features)
fit_col
```

```python
# Define X_train, y_train, and X_test
X = df[fit_col]
y = df['LogPrice']
X_test = df_test[fit_col]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X_test) # Linear Regression model makes a prediction
```

```python
df_test['SalePrice'] = np.exp(y_pred) # Prepare SalePrice format for submission
submission = df_test[['Id', 'SalePrice']] # Create submission dataset consisting of Id and SalePrice
```

```python
# Export submission as csv file
submission.to_csv('submission.csv', index=False)
```
